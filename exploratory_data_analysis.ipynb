### Load Required Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8')
```

### Load Cleaned Dataset

```python
df = pd.read_csv("cleaned_train_data.csv")

df.head()
```

### Basic Data Overview

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### Target Variable Analysis (Survived)

```python
df['Survived'].value_counts()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

#### ðŸ”¹ Fare Distribution

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Fare'])
plt.title("Fare Distribution")
plt.show()
```

#### ðŸ”¹ Gender Distribution

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution (0 = Female, 1 = Male)")
plt.show()
```

### Bivariate Analysis (Target vs Features)

#### ðŸ”¹ Survival vs Gender

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

#### ðŸ”¹ Survival vs Passenger Class

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

#### ðŸ”¹ Survival vs Age

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

### Multivariate Analysis

#### ðŸ”¹ Survival by Gender & Class

```python
sns.catplot(x='Pclass', y='Survived', hue='Sex', data=df, kind='bar')
plt.title("Survival by Class and Gender")
plt.show()
```

### Correlation Analysis

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Feature Relationship Exploration

#### ðŸ”¹ Family Size vs Survival

```python
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title("Survival Rate by Family Size")
plt.show()
```

### Pairplot (Optional but Powerful)

```python
sns.pairplot(df[['Age', 'Fare', 'FamilySize', 'Survived']], hue='Survived')
plt.show()
```

### Key EDA Conclusions 
### ðŸ” Key Insights from EDA
- Females had a much higher survival rate than males
- 1st class passengers were more likely to survive
- Higher fare correlated with higher survival probability
- Children and passengers with medium family sizes had better outcomes
